**roadNet-CA testing on Greedy SOSP and Robust MOSP**

```python
import time
from pathlib import Path
from scipy.io import mmread # required for .mtx files

import networkx as nx
from Greedy_SOSP import Greedy_SOSP
from Robust_MOSP import Robust_MOSP
```

```python
data_path = Path('data') / 'roadNet-CA.txt'
print('Dataset path:', data_path)
print('Exists:', data_path.exists())

G = nx.read_edgelist(
    data_path,
    create_using=nx.DiGraph(),
    nodetype=int,
    comments='#',
)

print('Loaded graph: nodes=', G.number_of_nodes(), 'edges=', G.number_of_edges())

# The full RoadNet-CA graph is large. Use a sampled subgraph for interactive testing if needed.
use_sample = True
sample_nodes = 100000

if use_sample:
    sample_nodes = min(sample_nodes, G.number_of_nodes())
    sample_node_list = list(G.nodes())[:sample_nodes]
    H = G.subgraph(sample_node_list).copy()
    print('Sampled graph: nodes=', H.number_of_nodes(),
          'edges=', H.number_of_edges())
else:
    H = G

G = H
```

```python
print('Graph summary:')
print('  nodes =', G.number_of_nodes())
print('  edges =', G.number_of_edges())
print('  sample edges =', list(G.edges())[:10])
```

```python
start = time.perf_counter()
energy, parent = Greedy_SOSP(G)
end = time.perf_counter()

print('Greedy_SOSP_Update completed')
print('  runtime =', end - start, 'seconds')
print('  sample energy entries =', list(energy.items())[:10])
print('  sample parent entries =', list(parent.items())[:10])
```

```python
start = time.perf_counter()
result = Robust_MOSP(G)
end = time.perf_counter()

print('Robust_MOSP_Update completed')
print('  runtime =', end - start, 'seconds')
print('  result type =', type(result))
print('  result keys =', list(result.keys()) if result is not None else result)

if result is not None:
    print('  path length =', len(result['path']))
    print('  robust_cost =', result['robust_cost'])
```
