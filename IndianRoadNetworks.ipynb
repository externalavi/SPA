```python
import networkx as nx

# Load the graph from the edgelist file as a directed graph
G_loaded = nx.read_edgelist(
    'data/roadNet-CA.txt',
    create_using=nx.DiGraph(),
    nodetype=int,
)

# Check counts
print("Directed graph:", G_loaded.is_directed())
print("Number of vertices:", G_loaded.number_of_nodes())
print("Number of edges:", G_loaded.number_of_edges())
```
